```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("output/summary/favela_metrics.csv")

# somente 2017 e 2020
df = df[df.year.isin([2017, 2020])].copy()

# ordena pelas maiores redes de 2017
order = (
    df[df.year == 2017]
    .sort_values("network_length_m", ascending=False)
    .favela
)

pivot = (
    df.pivot(
        index="favela",
        columns="year",
        values="network_length_m",
    )
    .loc[order]
)

ax = pivot.plot(
    kind="bar",
    figsize=(12,6),
    width=0.8,
)

ax.set_ylabel("Network length (m)")
ax.set_xlabel("")
ax.legend(title="")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
delta = (
    100
    * (pivot[2020] - pivot[2017])
    / pivot[2017]
)

delta.sort_values().plot.barh(
    figsize=(8,5)
)

plt.xlabel("Change in network length (%)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o CSV
df = pd.read_csv("output/summary/favela_metrics.csv")

# Comprimento da rede
pivot_length = df.pivot(
    index="favela",
    columns="year",
    values="network_length_m"
)

# Contexto vertical
pivot_context = df.pivot(
    index="favela",
    columns="year",
    values="context_height_median"
)

# Variação percentual da rede
delta_network = (
    (pivot_length[2020] - pivot_length[2017])
    / pivot_length[2017]
    * 100
)

# Contexto médio entre os dois anos
context = (
    pivot_context[2017] + pivot_context[2020]
) / 2

plot = pd.DataFrame({
    "context": context,
    "delta": delta_network,
})

plt.figure(figsize=(6,6))

plt.scatter(
    plot["context"],
    plot["delta"],
    s=80
)

for favela, row in plot.iterrows():
    plt.text(
        row["context"],
        row["delta"],
        favela,
        fontsize=8
    )

plt.xlabel("Median context height (m)")
plt.ylabel("Change in network length (%)")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carrega os dados
df = pd.read_csv("output/summary/favela_metrics.csv")

# Pivot das métricas
network = df.pivot(
    index="favela",
    columns="year",
    values="network_length_m"
)

context = df.pivot(
    index="favela",
    columns="year",
    values="context_height_median"
)

# Variação percentual
delta_network = (
    (network[2020] - network[2017])
    / network[2017]
) * 100

delta_context = (
    (context[2020] - context[2017])
    / context[2017]
) * 100

plot = pd.DataFrame({
    "delta_context": delta_context,
    "delta_network": delta_network,
})

# Correlação
r = plot.corr().iloc[0, 1]

# Gráfico
plt.figure(figsize=(7,6))

plt.scatter(
    plot["delta_context"],
    plot["delta_network"],
    s=80
)

for favela, row in plot.iterrows():
    plt.text(
        row["delta_context"] + 0.05,
        row["delta_network"] + 0.1,
        favela,
        fontsize=8
    )

plt.axhline(0, color="gray", lw=1)
plt.axvline(0, color="gray", lw=1)

plt.xlabel("Change in median context height (%)")
plt.ylabel("Change in network length (%)")
plt.title(f"Correlation = {r:.2f}")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
coef = np.polyfit(
    plot["delta_context"],
    plot["delta_network"],
    1
)

x = np.linspace(
    plot["delta_context"].min(),
    plot["delta_context"].max(),
    100
)

plt.plot(
    x,
    coef[0] * x + coef[1],
    color="red",
    lw=2
)
```

```python
from scipy.stats import pearsonr, spearmanr

# Remove Heliópolis
plot_sem_helio = plot.drop(index="heliópolis", errors="ignore")
plot_sem_helio = plot_sem_helio.drop(index="heliopolis", errors="ignore")

# Correlações
r, p = pearsonr(
    plot_sem_helio["delta_context"],
    plot_sem_helio["delta_network"]
)

rho, p_s = spearmanr(
    plot_sem_helio["delta_context"],
    plot_sem_helio["delta_network"]
)

print(f"Pearson : r = {r:.3f}   p = {p:.4f}")
print(f"Spearman: ρ = {rho:.3f}   p = {p_s:.4f}")

# Gráfico
plt.figure(figsize=(7,6))

plt.scatter(
    plot_sem_helio["delta_context"],
    plot_sem_helio["delta_network"],
    s=80
)

# reta de regressão
coef = np.polyfit(
    plot_sem_helio["delta_context"],
    plot_sem_helio["delta_network"],
    1
)

x = np.linspace(
    plot_sem_helio["delta_context"].min(),
    plot_sem_helio["delta_context"].max(),
    100
)

plt.plot(
    x,
    coef[0] * x + coef[1],
    color="red",
    lw=2
)

for favela, row in plot_sem_helio.iterrows():
    plt.text(
        row["delta_context"] + 0.2,
        row["delta_network"] + 0.1,
        favela,
        fontsize=8
    )

plt.xlabel("Change in median context height (%)")
plt.ylabel("Change in network length (%)")
plt.title(f"Without Heliópolis\nPearson={r:.2f}  Spearman={rho:.2f}")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
